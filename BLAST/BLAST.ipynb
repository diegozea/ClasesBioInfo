````python
from Bio.SeqUtils import GC

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

gc = np.zeros( len(seq) )
w = 21 # Ventana
c = 10 # Centro de la ventana: 11
for i in range(0, len(seq) - w + 1):
    gc[i+c] = GC(seq[i:i+w])

plt.plot(gc);
````

````python
def H(dna):
    """Entropia"""
    N = len(dna)
    suma = 0.0
    for letra in "ACTG":
        p = dna.count(letra) / float(N)
        if p > 0:
            suma += p * np.log2(p)
    return( -1.0 * suma )
````

````python
for i in range(1,5):
    sub = "ACTG"[0:i]
    print("%s\t%f" % ( sub, H(sub) ) )
````

````python
bits = np.zeros( len(seq) )
w = 41 # Ventana
c = 20 # Centro de la ventana: 21
for i in range(0, len(seq) - w + 1):
    bits[i+c] = H(seq[i:i+w])

plt.plot(bits);
````
