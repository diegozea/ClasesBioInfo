## Índice *Phred*  
Codificado en **FASTAQ** usando *ASCII*

$P_e$ = Probabilidad de base “erronea”  
$QV = -10· log(P_e)$  

Se considera aceptable hasta $QV=20$ (probabilidad de error del $0.01$)

```python
import numpy as np
for Pe in [0.001, 0.01, 0.1]:
    QV = -10 * np.log10( Pe )
    print( "QV = " + str( QV ) + "\t" + "Pe = " + str( Pe ) )
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from Bio import SeqIO
filename = "yeast_RNASeq_excerpt_sequence.txt"
for i,record in enumerate(SeqIO.parse(filename, "fastq-solexa")):
    plt.plot( [ SeqIO.QualityIO.phred_quality_from_solexa( q ) for q in record.letter_annotations["solexa_quality"] ] )
    if i == 50:
        break
plt.ylabel("PHRED quality score")
plt.xlabel("Position");
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
ftp = "ftp://ftp.ncbi.nlm.nih.gov/genomes/GENOME_REPORTS/"
pro = pd.read_table(ftp+"prokaryotes.txt",na_values=["-"],usecols=[6,14]).dropna()
euc = pd.read_table(ftp+"eukaryotes.txt",na_values=["-"], usecols=[6,14]).dropna()
vir = pd.read_table(ftp + "viruses.txt", na_values=["-"], usecols=[6,10]).dropna()

fig, coo = plt.subplots()
fig.set_figwidth(20); fig.set_figheight(10)
coo.plot(pro["Size (Mb)"], pro["Genes"], "ro", label="Procariota")
coo.plot(euc["Size (Mb)"], euc["Genes"], "bo", label="Eucariota")
coo.plot(vir["Size (Kb)"]/1000.0, vir["Genes"], "go", label="Virus")
coo.loglog(); coo.legend(); coo.set_xlabel("Size [Mb]"); coo.set_ylabel("Genes");
```

# Coverage (cobertura)

Se estima como el total de **pb secuenciadas / total de pb del genoma** a secuenciar  
Se puede entender como la cantidad de veces que una base es interrogada  

**Recomendación**  

- *Differential Gene Expression*: *Moderate* $10X$ ~ $30X$  
- *De novo Assembly*: $> 50X$   


```python
def cobertura(lecturas, longitud, genoma):
    """lecturas por biblioteca
    logitud de las lecturas en pb
    total de pb del genoma"""
    res = (lecturas*longitud) / genoma
    print( str(res) + "X" )
    return( res )
```

Arabidopsis: Transcriptoma 50Mb  
Lecturas por biblioteca: Promedio 15000000   
Longitud: 100 pb  
Profundidad:  

```python
cobertura(15000000,100,50e6);
```
