```python
endo = 0 # Contador para que los ID no sean iguales
exo  = 0
other= 0

import re # Expresion regular para buscar ENDO & EXO
buscar_endo = re.compile("endo.*polygalacturonase", flags=re.IGNORECASE) 
buscar_exo = re.compile("exo.*polygalacturonase", flags=re.IGNORECASE)

# Abro un archivo nuevo para escribir (w)
with open("new_id.fasta", "w") as fh_out:
    # Abro el archivo de BLAST para leer (r)
    with open("blast_aln.txt", "r") as fh_in:
        # Recorre cada linea de los FASTA del BLAST
        for line in fh_in:
            # Si es la linea de identificacion
            if line[0]==">":
                if buscar_endo.search(line):
                    endo += 1
                    # http://stackoverflow.com/questions/3228865/python-how-do-i-format-a-number-with-a-variable-number-of-digits
                    fh_out.write(">ENDO" + str(endo).zfill(6) + "|" + line[1:])
                elif buscar_exo.search(line):
                    exo += 1
                    fh_out.write(">EXO" + str(exo).zfill(7) + "|" + line[1:])
                else:
                    other += 1
                    fh_out.write(">OTHER" + str(other).zfill(5) + "|" + line[1:])
            else:
                fh_out.write(line)
            
```

```python
# Los valores negativos de distancias dan errores 
# Los busco linea por linea y los reemplazo por 0 en un nuevo archivo:
neg = re.compile("-\d+\.*\d+")

with open("positive_blast_tree","w") as fh_out:
    with open("blast_tree","r") as fh_in:
        for line in fh_in:
                # https://docs.python.org/2/library/re.html#re.sub
                fh_out.write(re.sub(neg, "0.00000", line))
```

```python
from Bio import Phylo
# http://biopython.org/wiki/Phylo
```

```python
arbol = Phylo.read("positive_blast_tree","newick")
```

```python
# Reemplazo los colores de las ramas, segun son ENDO, EXO o OTHER
for esp in arbol.get_terminals():
    if str(esp.name)[:4]=="ENDO":
        esp.color = "green"
    elif str(esp.name)[:3]=="EXO":
        esp.color = "yellow"
    else:
        esp.color = "red"
```
